```python
The mathematical formula for a linear Support Vector Machine (SVM) is expressed as follows:
Given a set of training samples (x,y), where x represents the feature vector of the i-th sample and y represents its corresponding class label
(-1 or 1 for binary classification), the goal of the linear SVM is to find the optimal hyperplane that separates the samples into different classes while maximizing the margin.

The decision function for a linear SVM is given by: f(x) = WtX + b where W is the weight vector that defines the orientation of the hyperplane
b is the bias term (also known as the intercept). X is the input feature vector.
During training, the SVM seeks to find the optimal values of 
�
w and 
�
b that maximize the margin (i.e., the distance between the hyperplane and the closest samples, known as support vectors) while still correctly classifying all training samples. This can be formulated as an optimization problem, typically solved using techniques like quadratic programming.
```

```python
Q2. What is the objective function of a linear SVM?
```

```python
The objective function of a linear Support Vector Machine (SVM) is formulated to maximize the margin between the decision boundary (hyperplane) and the closest data points (support vectors) while ensuring that all data points are correctly classified.
The objective function aims to minimize the L2 norm of the weight vector 
�
w, which corresponds to maximizing the margin between the decision boundary and the support vectors. The constraints ensure that all training samples are correctly classified with a margin of at least 1.

In essence, the objective function of a linear SVM seeks to find the optimal hyperplane that separates the classes while maximizing the margin and maintaining a good generalization ability. This formulation leads to a convex optimization problem, which can be efficiently solved using optimization techniques such as quadratic programming or gradient descent methods.
```

```python
Q3. What is the kernel trick in SVM?
```

```python
The kernel trick is a concept used in Support Vector Machines (SVMs) to extend linear classifiers to handle non-linear relationships between input features. It allows SVMs to effectively model complex decision boundaries in high-dimensional feature spaces without explicitly transforming the input features.

In SVMs, the decision boundary (hyperplane) is defined by a linear combination of the input features. However, in many real-world scenarios, the relationship between features and the target variable may not be linearly separable. To address this limitation, the kernel trick is employed, which involves mapping the input features into a higher-dimensional space using a kernel function.
```

```python
Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?
```

```python
Hyperplane:

The hyperplane is the decision boundary that separates the classes in a linear SVM.
In a two-dimensional feature space, the hyperplane is a line.
In a three-dimensional feature space, the hyperplane is a plane.
In higher-dimensional spaces, the hyperplane is a hyperplane.
Example: Consider a binary classification problem with two features (X1 and X2) and two classes (positive and negative). The hyperplane is a line that separates the positive and negative instances.
Marginal Plane:

The marginal plane is the plane parallel to the hyperplane that passes through the support vectors.
It defines the margin, which is the distance between the hyperplane and the support vectors.
Example: In a two-dimensional feature space, the marginal plane is the line that runs parallel to the hyperplane and passes through the support vectors.
Hard Margin:

Hard margin SVM aims to find the hyperplane that perfectly separates the classes without any misclassification.
It is suitable when the classes are linearly separable and there is no noise in the data.
Example: In a two-dimensional feature space, a hard margin SVM finds the hyperplane that perfectly separates the positive and negative instances without any overlap.
Soft Margin:

Soft margin SVM allows for some misclassification to achieve a balance between maximizing the margin and minimizing misclassification.
It is suitable when the classes are not perfectly separable or when there is noise in the data.
Example: In a two-dimensional feature space, a soft margin SVM finds the hyperplane that maximizes the margin while allowing for some misclassification of data points that lie within the margin.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for binary classification
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.where(X[:, 0] + X[:, 1] > 0, 1, -1)

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Positive Class')
plt.scatter(X[y == -1, 0], X[y == -1, 1], color='red', label='Negative Class')

# Plot the hyperplane
w = np.array([1, 1])  # Example weight vector
b = -0.5  # Example bias term
x_hyperplane = np.linspace(-2, 2, 100)
y_hyperplane = -(w[0] * x_hyperplane + b) / w[1]
plt.plot(x_hyperplane, y_hyperplane, color='black', linestyle='--', label='Hyperplane')

# Plot the margin
margin = 1 / np.linalg.norm(w)
y_margin_upper = y_hyperplane + margin / np.sqrt(np.sum(w ** 2))
y_margin_lower = y_hyperplane - margin / np.sqrt(np.sum(w ** 2))
plt.fill_between(x_hyperplane, y_margin_upper, y_margin_lower, color='yellow', alpha=0.3, label='Margin')

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Hyperplane, Margin, Soft Margin, and Hard Margin in SVM')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
This code generates random data for binary classification and visualizes the hyperplane, margin, soft margin, and hard margin in SVM. The hyperplane separates the positive and negative classes, while the margin defines the distance between the hyperplane and the support vectors. Soft margin and hard margin are illustrated by allowing for some misclassification within the margin or enforcing no misclassification, respectively.
```

```python
Q6. SVM Implementation through Iris dataset.
~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from matplotlib.colors import ListedColormap

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for visualization
y = iris.target

# Step 2: Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train a linear SVM classifier on the training set
svm_clf = SVC(kernel='linear')
svm_clf.fit(X_train, y_train)

# Step 4: Predict the labels for the testing set
y_pred = svm_clf.predict(X_test)

# Step 5: Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Step 6: Plot the decision boundaries of the trained model using two of the features
def plot_decision_boundary(X, y, clf):
    h = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF']), alpha=0.5)

    # Plot the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['red', 'green', 'blue']), edgecolors='k')

    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title('Decision Boundaries of Linear SVM')

plt.figure(figsize=(10, 6))
plot_decision_boundary(X_train, y_train, svm_clf)
plt.show()

# Step 7: Experiment with different values of the regularization parameter C
C_values = [0.1, 1, 10, 100]
for C in C_values:
    svm_clf = SVC(kernel='linear', C=C)
    svm_clf.fit(X_train, y_train)
    y_pred = svm_clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy (C={}): {:.2f}".format(C, accuracy))
```
